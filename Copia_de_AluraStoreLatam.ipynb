### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
import matplotlib.pyplot as plt

# Lista de DataFrames por tienda
tiendas_dfs = [tienda, tienda2, tienda3, tienda4]

# Enumerar tiendas con √≠ndices desde 1
ts = list(enumerate(tiendas_dfs, start=1))

# Calcular facturaci√≥n total por tienda
facturacion = [df['Precio'].sum() for _, df in ts]

# Nombres de las tiendas
nombres_tiendas = [f"Tienda {i}" for i, _ in ts]

# Mostrar facturaci√≥n por consola
for i, df in ts:
    total = df['Precio'].sum()
    print(f"Tienda {i}: $ {total:,.0f}")

# Gr√°fico de barras
plt.figure(figsize=(8, 6))
plt.bar(nombres_tiendas, facturacion, color='skyblue')
plt.title('Facturaci√≥n Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturaci√≥n ($)')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categor√≠a

```python
import matplotlib.pyplot as plt

# Lista de DataFrames por tienda
tiendas_dfs = [tienda, tienda2, tienda3, tienda4]
ts = list(enumerate(tiendas_dfs, start=1))

for i, df_tienda in ts:
    print(f"\nüìä Tienda {i} - Ventas por Categor√≠a")

    # Agrupar por categor√≠a del producto y contar cantidad de ventas
    ventas_por_categoria = (
        df_tienda.groupby('Categor√≠a del Producto')
                 .size()
                 .sort_values(ascending=False)
                 .reset_index(name='N√∫mero de Ventas')
    )
    print(ventas_por_categoria)

    # Conteo por producto dentro de cada categor√≠a
    conteo_productos = df_tienda.groupby('Categor√≠a del Producto')['Producto'].value_counts()
    print("\nüõí Ventas por producto dentro de cada categor√≠a:")
    print(conteo_productos)

    # Gr√°fico de barras
    plt.figure(figsize=(10, 6))
    plt.bar(ventas_por_categoria['Categor√≠a del Producto'], ventas_por_categoria['N√∫mero de Ventas'], color='mediumseagreen')
    plt.title(f"Ventas por Categor√≠a - Tienda {i}")
    plt.xlabel('Categor√≠a del Producto')
    plt.ylabel('N√∫mero de Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

# 3. Calificaci√≥n promedio de la tienda


```python
import matplotlib.pyplot as plt

# Lista de DataFrames por tienda
tiendas_dfs = [tienda, tienda2, tienda3, tienda4]
ts = list(enumerate(tiendas_dfs, start=1))

# Nombres de las tiendas
nombres_tiendas = [f"Tienda {i}" for i, _ in ts]

# C√°lculo de m√©tricas
facturacion = [df['Precio'].sum() for _, df in ts]
calificaciones = [df['Calificaci√≥n'].mean() for _, df in ts]

# Mostrar calificaciones por consola
print("\nCalificaci√≥n promedio por tienda:")
for nombre, cal in zip(nombres_tiendas, calificaciones):
    print(f"{nombre}: {cal:.2f}")

# Crear figura y ejes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Barras de facturaci√≥n
bars = ax1.bar(nombres_tiendas, facturacion, color='skyblue', label='Facturaci√≥n')
ax1.set_xlabel('Tienda')
ax1.set_ylabel('Facturaci√≥n ($)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Etiquetas sobre cada barra
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height, f"${height:,.0f}",
             ha='center', va='bottom', fontsize=9)

# L√≠nea de calificaci√≥n promedio
ax2 = ax1.twinx()
ax2.plot(nombres_tiendas, calificaciones, color='crimson', marker='o', linewidth=2, label='Calificaci√≥n Promedio')
ax2.set_ylabel('Calificaci√≥n Promedio', color='crimson')
ax2.tick_params(axis='y', labelcolor='crimson')
ax2.set_ylim(0, 5)

# T√≠tulo y dise√±o
plt.title('Facturaci√≥n y Calificaci√≥n Promedio por Tienda')
fig.tight_layout()
ax1.grid(True, axis='y', linestyle='--', alpha=0.3)

plt.show()
```

# 4. Productos m√°s y menos vendidos

```python
import matplotlib.pyplot as plt

# Lista de tiendas enumeradas
tiendas_dfs = [tienda, tienda2, tienda3, tienda4]
ts = list(enumerate(tiendas_dfs, start=1))

# Recorrer tiendas y graficar productos m√°s vendidos
for i, df in ts:
    productos_vendidos = df['Producto'].value_counts()

    print(f"\nüõí Tienda {i} - Productos vendidos:")
    print(productos_vendidos)

    # Gr√°fico horizontal
    plt.figure(figsize=(10, 6))
    productos_vendidos.sort_values().plot.barh(color='steelblue')
    plt.title(f"Tienda {i} - Productos M√°s Vendidos")
    plt.xlabel('Cantidad de Ventas')
    plt.ylabel('Producto')
    plt.grid(axis='x', linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

# 5. Env√≠o promedio por tienda

```python
import matplotlib.pyplot as plt

# Lista de tiendas enumeradas
tiendas_dfs = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = [f"Tienda {i+1}" for i in range(len(tiendas_dfs))]

# Calcular costos de env√≠o promedio
envios_promedio = [df['Costo de env√≠o'].mean() for df in tiendas_dfs]

# Mostrar los costos promedio de env√≠o
print("\n Costo de env√≠o promedio por tienda:")
for nombre, costo in zip(nombres_tiendas, envios_promedio):
    print(f"{nombre}: ${costo:,.2f}")

# Crear gr√°fico de pastel
plt.figure(figsize=(8, 8))

# Cambiar la paleta de colores con una mejor distribuci√≥n y colores agradables
colores = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']  # Colores c√°lidos y frescos

# Crear el gr√°fico de pastel
plt.pie(
    envios_promedio,
    labels=nombres_tiendas,
    autopct=lambda p: f'{p:.1f}%\n(${(p / 100) * sum(envios_promedio):,.0f})',
    startangle=140,
    colors=colores,
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}
)

# T√≠tulo y detalles del gr√°fico
plt.title('üì¶ Distribuci√≥n del Costo de Env√≠o Promedio por Tienda', fontsize=14)
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

```python
import plotly.graph_objects as go

# Datos de ejemplo (debes reemplazarlos con tus propios datos si ya los tienes)
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [120000, 95000, 87000, 65000]         # Facturaci√≥n por tienda
calificaciones = [4.5, 4.2, 3.9, 3.5]               # Calificaci√≥n promedio por tienda
envios_promedio = [26018.61, 25216.24, 24805.68, 23459.46]  # Costo de env√≠o promedio

# Lista de colores para cada tienda (uno por tienda)
colores = ['red', 'blue', 'green', 'orange']

# Crear el gr√°fico 3D interactivo
fig = go.Figure()

for i in range(len(tiendas)):
    fig.add_trace(go.Scatter3d(
        x=[facturacion[i]],
        y=[calificaciones[i]],
        z=[envios_promedio[i]],
        mode='markers',
        marker=dict(
            size=10,
            color=colores[i],
            opacity=0.8
        ),
        name=tiendas[i],
        text=[f"{tiendas[i]}<br>Facturaci√≥n: ${facturacion[i]:,.0f}<br>Calificaci√≥n: {calificaciones[i]}<br>Env√≠o: ${envios_promedio[i]:,.2f}"],
        hoverinfo='text'
    ))

# Configurar dise√±o del gr√°fico
fig.update_layout(
    title='üìä An√°lisis Comparativo de Tiendas en 3D',
    scene=dict(
        xaxis=dict(title='Facturaci√≥n'),
        yaxis=dict(title='Calificaci√≥n Promedio'),
        zaxis=dict(title='Costo de Env√≠o Promedio'),
    ),
    legend_title='Tiendas',
    showlegend=True,
    template='plotly_white'
)

fig.show()
```
